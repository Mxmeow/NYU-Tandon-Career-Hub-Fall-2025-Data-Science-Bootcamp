# HW Questions (Numpy):
Define two custom numpy arrays, say A and B. Generate two new numpy arrays by stacking A and B vertically and horizontally.

```python
import numpy as np
A = np.array([[1,2],[2,3],[6,7],[6,9],[4,2]]).T
B = np.array([[1,2,3,4,5],[5,4,3,2,1]])
```

```python
vertical_stack = np.vstack((A, B))
horizontal_stack = np.hstack((A, B))

print(vertical_stack)
print(horizontal_stack)
```

Find common elements between A and B. [Hint : Intersection of two sets]

```python
intersect = np.intersect1d(A,B)
print(intersect)
```

Extract all numbers from A which are within a specific range. eg between 5 and 10. [Hint: np.where() might be useful or boolean masks]

```python
mask = np.where((A >= 5) & (A <= 10), True, False)
masked = mask * A
print(masked)
```

Filter the rows of iris_2d that has petallength (3rd column) > 1.5 and sepallength (1st column) < 5.0

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_2d = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])
filter = iris_2d[(iris_2d[:, 2] > 1.5) & (iris_2d[:, 0] < 5.0)]
print(filter)
```

# HW Questions (Pandas):
From df filter the 'Manufacturer', 'Model' and 'Type' for every 20th row starting from 1st (row 0).

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
x = df.columns.tolist()
lookinFor = ['Manufacturer', 'Model', 'Type']
values = [x.index(name) for name in lookinFor if name in x]
for index in values:
    column_name = x[index]
    print(f'{column_name} has:')
    print(df[column_name][::20].reset_index(drop=True).to_string(index=False)) 
    print() 

```

Replace missing values in Min.Price and Max.Price columns with their respective mean.

    df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')

```python
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
x = df.columns.tolist()
lookinFor = ['Min.Price', 'Max.Price']
values = [x.index(name) for name in lookinFor if name in x]
for column in lookinFor:
    if column in df.columns:
        mean_value = df[column].mean()  
        df[column].fillna(mean_value, inplace=True)
for column in lookinFor:
    print(f'{column} has:')
    print(df[column])
    print()
```

How to get the rows of a dataframe with row sum > 100?

    df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))

```python
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
newdf = []

for row in range(len(df.values)):
    x = sum(df.values[row])
    if x > 100:
        newdf.append(df.values[row])
newdf = pd.DataFrame(newdf)
final_array = newdf.to_numpy()
print(df.values)
print(final_array)
```
